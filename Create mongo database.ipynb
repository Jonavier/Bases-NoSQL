```python
import pymongo
import pandas as pd
import json
import os
```

```python
consolidado = pd.DataFrame()

directorio_bases = r'.\Bases'

for i in os.listdir(directorio_bases):
    for j in os.listdir(directorio_bases + '\\' + i):
        fecha = directorio_bases + '\\' + i  + '\\' + j
        df = pd.read_csv(fecha) 
        df = df.loc[df["ARL"] != "TOTAL"]
        df = df.fillna(0)
        consolidado = pd.concat([consolidado, df], ignore_index=True)
```

```python
consolidado.columns = ['año', 'mes', 'departamento', 'municipio', 'sect_eco', 'act_eco', 'ARL', 'numero_empresas',
                       'participacion_numero_empresas', 'trabajadores_dependientes', 'trabajadores_independientes',
                       'total_trabajadores', 'participacion_numero_trabajadores', 'accidentes_trabajo', 
                       'enfermedad_laboral', 'muertes_accidentes_trabajo', 'muertes_enfermedad_laboral', 'total_muertes',
                       'pensiones_accidentes_trabajo', 'pensiones_enfermedad_laboral', 'total_pensiones', 
                       'indemnizaciones_pagadas_accidentes_trabajo', 'indemnizaciones_pagadas_enfermedad_laboral',
                       'total_indemnizaciones_pagadas']
```
