```python
import pymongo
import pandas as pd
import json
import os
```

```python
consolidado = pd.DataFrame()

directorio_bases = r'.\Bases'

for i in os.listdir(directorio_bases):
    for j in os.listdir(directorio_bases + '\\' + i):
        fecha = directorio_bases + '\\' + i  + '\\' + j
        df = pd.read_csv(fecha) 
        df = df.loc[df["ARL"] != "TOTAL"]
        df = df.fillna(0)
        consolidado = pd.concat([consolidado, df], ignore_index=True)
```
